```python
from easygui import *
from selenium import webdriver

import requests
import getpass
from bs4 import BeautifulSoup
import re

print(requests.__version__)
print(bs4.__version__)
print(re.__version__)
```

```python
from robobrowser import RoboBrowser
usr = input("Enter Your Username : ")
pas = getpass.getpass("Enter your password : ")
br = RoboBrowser()
br.open("http://classroom.dwit.edu.np/login/index.php")
form = br.get_form()
form['username'] = usr
form['password'] = pas
br.submit_form(form)
```

```python
src = str(br.parsed())
html = BeautifulSoup(src, 'html.parser')
print(src)
```

```python
course_section = html.find(id = 'frontpage-course-list')
print(course_section)
```

```python
courses_and_teachers = course_section.select("a")
print(courses_and_teachers)
```

```python
courses_only = []
for link in courses_and_teachers:
    if re.findall('course/',link.attrs['href']) :
            courses_only.append(link)
print(courses_only)
```

```python
for link in courses_only:
    print(link.text + " => " + link.attrs['href'])
```

```python
for link in courses_only:
    print(link.text)
sub = input("Subject Code to scrap data : ")
```

```python
for link in courses_only:
    if re.findall(sub, link.text):
        sub_url = link.attrs['href']
print(sub_url)
```

```python
br.open(sub_url)
src = str(br.parsed())
print(src)
```

```python
sub_html = BeautifulSoup(src, 'html.parser')
print(sub_html)
```

```python
assign_divs = sub_html.findAll("div", {"class": "activityinstance"})
```

```python
assign_temp_a = []
i=1
for div in assign_divs:
    if not re.findall('Assignment', div.text):
        i -=1
        del assign_divs[i]
    i+=1
print(type(assign_divs))
```

```python
link = []
for j in range(i):
    assign_a = assign_divs[j].select('a')
    link.append(assign_a[0].attrs['href'])
print(link)
       
```

```python
i=0
for j in range(len(link)):
    br.open(link[j])
    pdf_file = 'assignment'

    content = br.response.content

    with open('./Assignments/' + pdf_file + str(i) + '.pdf', "wb") as output:
      output.write(content)
    i+=1
    
```
